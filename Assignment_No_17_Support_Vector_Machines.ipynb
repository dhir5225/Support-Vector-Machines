# SVM

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
```

# Forest

```python
forest=pd.read_csv('forestfires.csv')
forest
```

```python
forest=forest.iloc[:,2:]
forest
```

```python
lb=LabelEncoder()
forest['size_category']=lb.fit_transform(forest['size_category'])
```

```python
forest['size_category'].unique()
```

```python
forest
```

```python
X=forest.iloc[:,0:28]
Y=forest['size_category']
```

```python
X
```

```python
Y
```

```python
forest_df=pd.DataFrame(X,columns=forest.columns[0:28])
forest_df
```

```python
print('\n + The unique labels in this data are'+ str(np.unique(Y)))
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,shuffle=True)
```

```python
print('There are {} in training set & {} in test'.format(X_train.shape[0],X_test.shape[0]))
print()
```

```python
sc=StandardScaler()
sc.fit(X_train)
```

```python
X_train_std=sc.transform(X_train)
X_test_std=sc.transform(X_test)
```

```python
print(pd.DataFrame(X_train_std,columns=forest_df.columns))
```

```python
svm=SVC(kernel='rbf',random_state=4,gamma=0.10,C=1.0)
svm.fit(X_train,Y_train)
```

```python
svm.score(X_train_std,Y_train)
```

```python
svm.score(X_test_std,Y_test)
```
